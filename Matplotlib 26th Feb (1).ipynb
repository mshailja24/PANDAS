```python
x=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
y=[5,6.7,4,6,2,4.9,1.8]
```

```python
a=sns.stripplot(x,y)
a.set(xlabel='Days',ylabel='Amount_spend')
plt.title('My_first_graph')
plt.show()
```
