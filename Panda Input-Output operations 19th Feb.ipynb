```python
d=pd.to_datetime({'year':[2015],'month':[2],'day':[4]})
d                  
```

```python
#inside pandas to_datetime functions are present
df['Hiredate']=pd.to_datetime(df['Hiredate'])
df
```

```python
#pandas.date_range
```

```python
#date formate-----.mm/dd/yyyy
daterange=pd.date_range(start='1/1/2020',end='1/10/2020',freq='D')
daterange
```

```python
daterange=pd.date_range(start='1/1/2020',end='5/1/2020',freq='M')
daterange
```

```python
import datetime
dt=datetime.datetime.now()
print(dt)

```

```python
#timedelta----> Represents a duration,the difference b/w two dates or times.
print(dt+ pd.to_timedelta(10,unit='D'))
```

```python
#adding new column day,month,year in dataframe with dt attribute

df['Day']=df['Hiredate'].dt.day
df['Month']=df['Hiredate'].dt.month
df['Year']=df['Hiredate'].dt.year
print(df)
```

```python
import pandas as pd
df=pd.read_csv('empl.csv')
print(df)
```

```python
#same statement can be written in MYSQL database by select name,age,salary from table/(here dataframe);
df1=pd.DataFrame(data=df,columns=['Name','Age','Salary'])
df1
```

```python
#select name,age,salary from hrdata Limit 5;
df[['Name','Age','Salary']].head(5)
```

```python
#select * from hrdata/df where City=='newyork'
df[df["City"]=="Newyork"]
```

```python
#select * from df/hrdata WHERE City='newyork' AND Salary>50000;
df[(df["City"]=="newyork")&(df["Salary"]>50000)]
```

```python
#select * from df/hrdata WHERE City= Africa or Salary>65000
df[(df["Country"]=="Africa")| (df["Salary"]>65000)]
```

```python
#SELECT * FROM df WHERE column IS NULL;
df[df["Salary"].isna()]
```

```python
#SELECT * FROM frame WHERE col1 IS NOT NULL;
df[df["Salary"].notna()]
```

```python
#axis=0----->signifies rows----default
#df5=df.dropna()
#df.dropna(axis=0,inplace=True)-----will take changes in original dataframe

df5=df.dropna(axis=0)
print(df5)
```
